```python
"""
1. Grocery Billing
A shopkeeper wants to calculate the total bill for two items. Write a program that takes
the price and quantity of each item and prints the total bill amount.
Sample Input:
Item1 price: 45
Item1 quantity: 3
Item2 price: 30
Item2 quantity: 2
Sample Output:
Total Bill: 195
"""
Item1 = int(input("Enter the price of first item: "))
Item1_quantity = int(input("Enter the quality of first item: "))
Item2 = int(input("Enter the price of second item: "))
Item2_quantity = int(input("Enter the quality of second item: "))
Item1_price = Item1 * Item1_quantity
Item2_price = Item2 * Item2_quantity
Total_bill = Item1_price + Item2_price
print("Total bill: ", Total_bill)
```

```python
"""
2. Temperature Conversion
Write a program to convert temperature from Celsius to Fahrenheit using the formula ¬∞F
= (¬∞C √ó 9/5) + 32.
Sample Input:
Enter temperature in Celsius: 30
Sample Output:
Temperature in Fahrenheit: 86.0
"""
Celsius = int(input("Enter temperature in Celsius: "))
Fahrenheit = (Celsius * 9/5) + 32
print("Temperature in Fahrenheit: ", Fahrenheit)
```

```python
"""
3. Salary Slip
Given the basic salary of an employee, calculate the total salary including HRA (20%) and
DA (10%).
Sample Input:
Enter basic salary: 20000
Sample Output:
Net Salary: 26000.0
"""
basic_salary = float(input("Enter basic salary:"))
hra = basic_salary * 0.2
da = basic_salary * 0.1
total_salary = basic_salary + hra + da
print("Net Salary:", total_salary)
```

```python
"""
4. Simple Interest
Write a program to calculate simple interest using the formula SI = (P √ó R √ó T) / 100.
Sample Input:
Principal: 10000
Rate of Interest: 5
Time (in years): 2
Sample Output:
Simple Interest = 1000.0
"""
P = int(input("Enter Principal amount: "))
R = int(input("Enter Rate of Interest: "))
T = int(input("Enter Time in years: "))
SI = (P * R * T) / 100
print("Simple Interest = ", SI)
```

```python
"""
5. Time Conversion
Write a program to convert total minutes into hours and minutes.
Sample Input:
Enter total minutes: 130
Sample Output:
2 hours and 10 minutes
"""
total_minutes = int(input("Enter total minutes: "))
hours = total_minutes // 60
minutes = total_minutes % 60
print(hours, "hours and", minutes, "minutes")
```

```python
"""
6. BMI Calculator
Write a program to calculate the Body Mass Index (BMI) using the formula: BMI
weight(kg) / (height(m) √ó height(m))
Sample Input:
Weight: 60
Height: 1.65
Sample Output:
BMI = 22.04
"""
weight = float(input("Enter weight in kg: "))
height = float(input("Enter height in m: "))
BMI = weight / (height * height)
print(f"BMI = {BMI:.2f}")
```

```python
"""
7. Discount Calculation
A store offers a 10% discount on every item. Write a program that takes the original price
and prints the discounted price.
Sample Input:
Original price: 1200
Sample Output:
Discounted Price: 1080.0
"""
price = float(input("Original price : "))
discount = price * 0.1
discounted_price = price - discount
print("Discounted Price: ", discounted_price)
```

```python
"""
8. Area of a Triangle (Using Heron‚Äôs Formula)
A land surveyor wants to calculate the area of an irregular triangular plot.
Write a Python program to find the area of a triangle when all three sides are known,
using Heron‚Äôs formula:
ùë† =
ùëé + ùëè + ùëê
2
Area = ‚àöùë†(ùë† ‚àí ùëé)(ùë† ‚àí ùëè)(ùë† ‚àí ùëê)
Sample Input:
Enter side a: 5
Enter side b: 6
Enter side c: 7
Sample Output:
Area of the triangle: 14.70
"""
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))
s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
print(f"Area of the triangle: {area:.2f}")
```

```python
"""
9. Paint Cost Estimator
A painter charges ‚Çπ25 per square meter. Write a program to calculate the total cost to
paint a rectangular wall. The wall has length and height, and there are two windows each
measuring 1.5m √ó 1.2m that should not be painted. Find the total area to be painted and
its total cost.
Sample Input:
Enter wall length (in meters): 12
Enter wall height (in meters): 8
Sample Output:
Total area to be painted: 92.4 sq.m
Total painting cost: ‚Çπ2310.0
"""
length = float(input("Enter wall length (in meters): "))
height = float(input("Enter wall height (in meters): "))
window_area = 1.5 * 1.2
wall_area = length * height
total_window_area = 2 * window_area
paint = wall_area - total_window_area
cost = paint * 25
print("Total area to be painted: ", paint, "sq.m")
print("Total painting cost: ‚Çπ", cost)
```

```python
"""
10. Electricity Bill Calculator
An electricity board charges consumers based on unit consumption. The rate per unit is
‚Çπ5.2. Additionally, there is a fixed charge of ‚Çπ100 on every bill. Write a Python program
that takes the number of units consumed and calculates the total bill amount.
Sample Input:
Enter number of units consumed: 250
Sample Output:
Electricity Bill: ‚Çπ1400.0
"""
units = int(input("Enter number of units consumed: "))
rate = 5.2
fixed_charge = 100
total_bill = (units * rate) + fixed_charge
print("Electricity Bill: ‚Çπ", total_bill)

```
