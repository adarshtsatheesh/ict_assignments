```python
# 1 For a given range of numbers (e.g., 1 to 100), print "Fizz" if the number is divisible by 3, "Buzz" if it's divisible by 5, and "FizzBuzz" if it's divisible by both. Otherwise, print the number itself

for i in range(1, 101):
  if i % 3 == 0 and i % 5 == 0:
    print("FizzBuzz")
  elif i % 3 == 0:
    print("Fizz")
  elif i % 5 == 0:
    print("Buzz")
  else:
    print(i)
```

```python
''' 2 Create a bill calculator where the user enters the number of units consumed. ₹5 per unit for the first 100 units ₹7 per unit for the next 100 units (from 101 to 200) ₹10 per
unit for any units above 200 Calculate and display the total electricity bill.'''

u=int(input("Enter the number of units consumed: "))
t=0
if u<=100:
  t=u*5
elif u<=200:
  t=(100*5)+((u-100)*7)
else:
  t=(100*5)+(100*7)+((u-200)*10)
print("Total electricity bill:",t)
```

```python
# 4 Develop a Python program that checks whether a given word or phrase is a palindrome
num=input('Enter num or string to be checked:')
if num==num[::-1]:
  print(num,'is a palindrome')
else:
  print(num,'is not a palindrome')
```
